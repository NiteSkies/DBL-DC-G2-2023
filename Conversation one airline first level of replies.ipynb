```python
#The idea here is to get an overview of the conversations with only replies from the first level
# Create a copy of the original DataFrame
df_conversations_first_reply_Lufthansa = df_all_tweets_Lufthansa.copy()

# Reset the index of the DataFrame
df_conversations_first_reply_Lufthansa.reset_index(drop=True, inplace=True)

# Create a dictionary to store the conversations
conversations = {}

# Iterate through the DataFrame rows
for _, row in df_conversations_first_reply_Lufthansa.iterrows():
    if pd.isnull(row['in_reply_to_status_id_str']):
        # New conversation starts
        conversation_id = row['id_str']
        conversations[conversation_id] = []
        conversations[conversation_id].append(row)
    else:
        # Reply in an existing conversation
        parent_tweet_id = row['in_reply_to_status_id_str']
        if parent_tweet_id in conversations:
            conversations[parent_tweet_id].append(row)

# Flatten the conversation dictionary into a list
conversation_list = [tweet for sublist in conversations.values() for tweet in sublist]

# Create a new DataFrame with the conversations
df_conversations_first_reply = pd.DataFrame(conversation_list)

# Create group index and reply index columns
df_conversations_first_reply_Lufthansa['group_index'] = df_conversations_first_reply_Lufthansa['in_reply_to_status_id_str'].isnull().cumsum()
df_conversations_first_reply_Lufthansa['reply_index'] = df_conversations_first_reply_Lufthansa.groupby('group_index').cumcount() + 1

# Set the two indexes for the DataFrame
df_conversations_first_reply_Lufthansa.set_index(['group_index', 'reply_index'], inplace=True)
```
