```python
import json
import pandas as pd

# Load the JSON file as a string
with open('airlines-1558527599826.json', 'r') as f:    #only the first JSON file 
    json_str = f.read()

# Split the string into a list of JSON objects
json_objs = json_str.split('\n')

# Parse each JSON object in the list and append to a list of dictionaries
dict_list = []
for json_obj in json_objs:
    if json_obj.strip() == '':
        continue
    dict_list.append(json.loads(json_obj))

# Convert the list of dictionaries to a Pandas DataFrame
df = pd.DataFrame(dict_list)
df


```
